```python

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
csv_path = r'f:\sv_mochoujiedao\色彩丰富度-cri.csv'

# 读取CSV文件
df = pd.read_csv(csv_path)
```

```python
df.head()
```

```python

# 创建归一化器
scaler = MinMaxScaler(feature_range=(0, 1))

# 归一化第二列数据
df['cri'] = scaler.fit_transform(df[df.columns[1]].values.reshape(-1, 1))

# 保存归一化后的数据到新的CSV文件
df.to_csv('f:\sv_mochoujiedao\色彩丰富度-01.csv', index=False)
```
