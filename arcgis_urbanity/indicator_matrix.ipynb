```python
import os  
import glob  
  
# 设置文件夹路径  
folder_path = r'E:\work\spatio_evo_urbanvisenv_svi\sv\perception_score'  # 请替换为你的文件夹路径  
# 使用glob模块搜索CSV文件  
csv_files = glob.glob(os.path.join(folder_path, '*.csv'))  
# 打印找到的CSV文件列表  
for csv_file in csv_files:  
    print(csv_file)
```

```python
import pandas as pd  

dfs = []
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    df.set_index(df.columns[0], inplace=True)  
    dfs.append(df)
  
# 使用pd.concat()函数合并DataFrame，这里使用axis=1来按列合并  
combined = pd.concat(dfs, axis=1)
```

```python
# combined = combined.reset_index(name='iamge_name') 
combined = combined.reset_index() 
combined.head()
# combined.columns
```

```python
# 提取列索引  
column_index = combined.columns.tolist()  
column_index
```

```python
combined.to_csv(r'E:\work\spatio_evo_urbanvisenv_svi\sv\perception_score\combined.csv', index=False)
```

```python

file_path1 = r'e:\work\spatio_evo_urbanvisenv_svi\sv\ss.csv'  # 替换为你的第一个CSV文件的路径  
# df1 = pd.read_csv(file_path1, usecols=[0, 2, 3,1,5,18,67,95,14,88,7,53,12,33,49,137])
df1 = pd.read_csv(file_path1)

# 设置第一列为索引值  
df1.set_index(df1.columns[0], inplace=True)
```
