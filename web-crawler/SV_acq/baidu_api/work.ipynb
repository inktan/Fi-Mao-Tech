```python
import pandas as pd
import os

# 设定工作目录，即存放CSV文件的目录
# os.chdir('path/to/your/csv/files')

# 找到所有CSV文件
# csv_files = [file for file in os.listdir() if file.endswith('.csv')]
csv_files = [
    r'f:\shanghaijiaoda_poi_shp\20240614\result\work03_id_address_anjuke.csv',
    r'f:\shanghaijiaoda_poi_shp\20240614\result\work03_id_公交站.csv',
    r'f:\shanghaijiaoda_poi_shp\20240614\result\work03_id_sport.csv',
    r'f:\shanghaijiaoda_poi_shp\20240614\result\work03_id_医院_诊所.csv',
    r'f:\shanghaijiaoda_poi_shp\20240614\result\work03_id_公园_广场.csv',
    r'f:\shanghaijiaoda_poi_shp\20240614\result\work03_id_商场.csv',
    r'f:\shanghaijiaoda_poi_shp\20240614\result\work03_id_rate.csv',
    r'f:\shanghaijiaoda_poi_shp\20240614\mixed_score\work03_id_land_use_mixed_score_all.csv',
             ]

# 读取第一个CSV文件作为基础
merged_df = pd.read_csv(csv_files[0])
merged_df = merged_df.drop_duplicates(subset='id')

# 循环遍历剩余的CSV文件，并按照id列合并
for file in csv_files[1:]:
    print('Merging file:', file)
    df = pd.read_csv(file)
    df = df.drop_duplicates(subset='id')

    merged_df = pd.merge(merged_df, df, on='id', how='outer')
    # merged_df = pd.merge(merged_df, df, on='id', how='inner')
    # merged_df = pd.concat([merged_df, df], ignore_index=True)

# 保存合并后的CSV文件
merged_df = merged_df.drop_duplicates(subset='id')
merged_df.to_csv(r'f:\shanghaijiaoda_poi_shp\20240614\result\work03_id_all.csv', encoding='utf-8', index=False)
```

```python
import pandas as pd  
  
# 定义CSV文件路径  
csv_path = r'f:\2022poi\2022POI\华东poi1\上海市\上海市POI数据.csv'  
df = pd.read_csv(csv_path)  
  
# 第一个条件：名称中包含'故居'、'博物馆'或'纪念馆'  
mask1 = df['名称'].str.contains('故居|博物馆|纪念馆', na=False, case=False)  
  
# 第二个条件：名称中不包含'停车场'  
mask2 = ~df['名称'].str.contains('停车场|地铁站', na=False, case=False)  # 使用~取反操作  
  
# 使用逻辑与（&）结合两个条件  
combined_mask = mask1 & mask2  
  
# 使用这个复合布尔掩码来过滤DataFrame  
filtered_df = df[combined_mask]  
  
# 定义输出文件路径  
output_file = r'f:\2022poi\2022POI\华东poi1\上海市\上海市POI数据_故居_博物馆_纪念馆_no_parking.csv'  
  
# 将过滤后的DataFrame保存到新的CSV文件中  
filtered_df.to_csv(output_file, index=False)  
  
# 输出filtered_df（这一步在脚本运行时通常不需要，因为它在控制台显示数据，  
# 但如果你想在脚本运行完毕后查看filtered_df的内容，保留这一步是有用的）  
filtered_df
```
